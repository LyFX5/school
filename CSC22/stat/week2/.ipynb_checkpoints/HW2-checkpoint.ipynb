```python

# Задача 4.

phi = EX

Norm = ss.norm(loc = 0, scale = 1)

R = Norm.ppf(0.975)
L = Norm.ppf(0.025)

P1 = 0
P2 = 0

N = 1000000
n = 138

for _ in range(N):
    
    sample = []
    for i in range(n):
        sample.append(play()[1])

    sample = np.array(sample)
    
    phi_star = sample.mean()

    D_phi_star = (sample**2).mean() - phi_star

    s_phi_star = D_phi_star**0.5
    
    if phi < phi_star - R * (s_phi_star/(n**0.5)):
        P1 += 1
        
    if phi > phi_star - L * (s_phi_star/(n**0.5)):
        P2 += 1
        
        
P1 /= N
P2 /= N

print(P1)
print(P2)
```

```python

# Задача 3.

def germ():
    
    X = dict.fromkeys(list(set(list(REWARDS.values()))), 0)
    
    P_eval = dict.fromkeys([0, 1, 2, 3, 5, 6, 7], 0)

    N = 138
    sample = []
    for i in range(N):
        comb = play()[0]
        for j in range(3):
            P_eval[comb[j]] += 1/(3*138)

    def prob(r):
        (a, b, c) = r
        return P_eval[a]*P_eval[b]*P_eval[c]

    def comb_num(r):
        return COMBS[r]


    for r in REWARDS:
        X[REWARDS[r]] += prob(r) * (comb_num(r))

    EX_germ = 0

    for r in X:
        EX_germ += X[r] * r

    return EX_germ

    
MSE_germ = 0

for j in range(10000):
    
    EX_germ = germ()
    
    MSE_germ += (EX_germ - EX)**2

print( (MSE_germ / 10000)**0.5 )
    
```
