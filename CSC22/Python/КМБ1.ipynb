```python
# MENU


def form(num):
    
    if len(num) <= 3:
        return num
    
    else:
        
        newNum = ''

        n = len(num)

        for i in range(1, n+1):

            dig = num[n-i]

            newNum = dig + newNum

            if i % 3 == 0 and i != n:
                newNum = ' ' + newNum

        return newNum



def exchange(price):

    dolar = ''
    
    end = ''

    for char in price:
        if char.isnumeric():
            dolar += char
        elif char != '$':
            end += char
    
    rub = str(int(dolar) * 75)
    
    return form(rub) + 'р' + end
    
import sys
    
for line in sys.stdin:
    menu = line.rstrip().split()
    
    new_menu = ''
    
    for i in range(len(menu)):

        if menu[i][0] == '$':
            new_menu += exchange(menu[i])

        else:
            new_menu += menu[i]

        if i != len(menu)-1:
            new_menu += ' '


    print(new_menu)

```

```python


def get_divisors(N):
    
    d = 2
    
    divisors = []

    while d**2 <= N:

        if N % d == 0:
            divisors.append(d)
            
        d += 1
            
    divisors1 = divisors.copy()
    
    n = len(divisors1)
            
    for i in range(1, n+1):
        
        d = divisors1[n-i]
        
        if d**2 != N:
            divisors.append(N // d)
        
    return divisors

divisors = sorted(get_divisors(int(input())))

if len(divisors) != 0:
    print(*divisors)
     
```

```python
# Напишите программу, которая умеет склонять с любыми числительными слово «комментарий». 

def inclination(n):
    
    I = ['комментарий', 'комментария', 'комментариев']
    
    if 0 < abs(n)%1 < 1:
        return I[1]
    
    else:
        
        n = n % 100
        
        if 10 <= n <= 20:
            return I[2]
        
        else:
            
            n = n % 10
            
            if n == 1:
                return I[0]
            
            elif n == 2 or n == 3 or n == 4:
                return I[1]
            
            else:
                return I[2]
                
print(inclination(float(input())))

```
