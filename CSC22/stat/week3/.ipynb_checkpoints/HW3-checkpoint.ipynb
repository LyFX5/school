```python

# Задача 2.

import random as rn

rewards = []
outputs = [] # эмпирическое распределение

file_sample = open('slot_138.txt', 'r')

for line in file_sample.readlines():
    [c1, c2, c3] = map(int, line.strip().split())
    
    output = tuple(sorted([c1, c2, c3]))
    
    reward = REWARDS[output]
    
    rewards.append(reward)
    outputs.append(output)
        
file_sample.close()


def bootstrap_generator_ron():
    
    output = rn.choice(outputs)
    
    reward = REWARDS[output]
    
    return [output, reward]


bootstrap_true_EX_Ron = sum(rewards) / len(rewards)

N = 100_000

n = len(rewards)

squared_errors = []

for i in range(N):
    bootstrap_sample = []
    
    for j in range(n):
        
        bootstrap_sample.append(bootstrap_generator_ron()[1])
        
    phi_star_bootstrap = sum(bootstrap_sample) / len(bootstrap_sample)
    
    squared_errors.append((phi_star_bootstrap - bootstrap_true_EX_Ron)**2)
    
RMSE_Ron = (sum(squared_errors) / len(squared_errors))**0.5


print(RMSE_Ron)

```

```python

# Задача 3.

import random as rn

import numpy as np
import scipy.stats as ss

outputs = [] # эмпирическое распределение

file_sample = open('slot_138.txt', 'r')

for line in file_sample.readlines():
    [c1, c2, c3] = map(int, line.strip().split())
    
    output = tuple(sorted([c1, c2, c3]))
    
    outputs.append(output)
    
file_sample.close()
    
    
n = len(outputs) # 138


def germiona_evaluation(sample):

    Approx_Distr_of_pictures = dict.fromkeys([0, 1, 2, 3, 5, 6, 7], 0)

    for output in sample: # n

        for picture in output: # 3

            Approx_Distr_of_pictures[picture] += 1/(3*n)


    def prob(r):
        (a, b, c) = r
        return Approx_Distr_of_pictures[a] * Approx_Distr_of_pictures[b] * Approx_Distr_of_pictures[c]


    Distribution_of_rewards = dict.fromkeys(list(set(list(REWARDS.values()))), 0)  
    

    for output in REWARDS:
        Distribution_of_rewards[REWARDS[output]] += prob(output) * (NUMBER_OF_PERTURBATIONS[output]) 
        # вероятность того что торойка этих картинок 
        # может появится в каком-то порядке на количество порядков

    bootstrap_EX_Germ = 0

    for reward in Distribution_of_rewards:
        bootstrap_EX_Germ += Distribution_of_rewards[reward] * reward


    return bootstrap_EX_Germ

bootstrap_true_EX_Germ = germiona_evaluation(outputs)


# create generator (аппроксимация истинного распределения) ==================

Approx_Distr_of_pictures = dict.fromkeys([0, 1, 2, 3, 5, 6, 7], 0)

for output in outputs: # n

    for picture in output: # 3

        Approx_Distr_of_pictures[picture] += 1/(3*n)


Approx_Distr_of_pictures = ss.rv_discrete(values=(list(Approx_Distr_of_pictures.keys()), list(Approx_Distr_of_pictures.values())))


def bootstrap_generator_germiona():
    
    sample = Approx_Distr_of_pictures.rvs(size=(3, n))
    
    return [tuple(sorted(x)) for x in zip(*sample)]

# create generator (аппроксимация истинного распределения) ==================


N = 100_000

squared_errors = []

for i in range(N):
    
    bootstrap_sample = bootstrap_generator_germiona()
    
    bootstrap_EX_Germ = germiona_evaluation(bootstrap_sample)
    
    squared_errors.append( (bootstrap_EX_Germ - bootstrap_true_EX_Germ)**2 )
    
RMSE_Germ = (sum(squared_errors) / len(squared_errors))**0.5

print(RMSE_Germ)

```
