```python

LA = ['A', 'B', 'C']

import random as rn

def choice_able(disabled):
    LA_able = LA.copy()
    for i in range(len(disabled)):
        lat = disabled[i]
        LA_able.remove(lat)
    
    #print(LA_able)
    choi = rn.choice(LA_able)
    return choi

n = len(LA)

L_square = []

disabled_in_row = []
disabled_in_column = []

for i in range(n):
    disabled_in_row.append([])
    disabled_in_column.append([])
    
dengerous_letters = []

i = 0

while i < n:
    row = []
    j = 0
    while j < n:
        
        disabled = list(set(disabled_in_row[i] + disabled_in_column[j] + dengerous_letters))
        dengerous_letters = []
        
        if len(disabled) == n:
            #  нужно вернуться на предыдущий шаг и запретить такой расклад
            
            dengerous_letters.append(disabled_in_column[j-1].pop())
            
            disabled_in_row[i].pop()
            
            row.pop()
            
            j -= 2
            
            #print("backtracking")
        
        else:
        
            lat = choice_able(disabled)

            row.append(lat)

            disabled_in_row[i].append(lat)
            disabled_in_column[j].append(lat)

        #print('===========')
        #print(disabled_in_row)
        #print(disabled_in_column)
        #print()
        
        j += 1

    L_square.append(row)
    
    i += 1
    
    

for i in range(n):
    for j in range(n):
        print(L_square[i][j], end=' ')
    print()

    
```

```python

# sudoku

Alpha = ['1','2','3','4']

n = len(Alpha)
 
import random as rn

reserved_in_square = {0: set(), 1: set(), 2: set(), 3: set()}

def number_of_square(point):
    
    [i, j] = point
    
    if (i+1)**2 <= n and (j+1)**2 <= n:
        # 1 sq
        return 0
            
    elif (i+1)**2 <= n and (j+1)**2 > n:
        # 2 sq
        return 1
        
    elif (i+1)**2 > n and (j+1)**2 <= n:
        # 3 sq
        return 2
        
    elif (i+1)**2 > n and (j+1)**2 > n:
        # 4 sq
        return 3
    else:
        print("wrong argument")
        return


reserved_in_row = []
reserved_in_column = []

for i in range(n):
    row1 = []
    row2 = []
    for j in range(n):
        row1.append('')
        row2.append('')
        
    reserved_in_row.append(row1)
    reserved_in_column.append(row2)
    
    
for i in range(n):
    line = input()
    
    for j in range(n):
        ch = line[j]
        
        if ch != '0':
            reserved_in_row[i][j] = ch
            reserved_in_column[j][i] = ch
            
            reserved_in_square[number_of_square([i, j])].add(ch)
            
            
            
permitted_points = []

for i in range(n):
    for j in range(n):
        if reserved_in_row[i][j] == '':
            permitted_points.append([i, j])


class cell:
    def __init__(self):
        self.point = [0, -1]
        self.value = ''
        self.posible_steps = Alpha.copy()
        self.kids = []
        
        self.parent = None
        
        self.visited = False
        
        self.square_number = -1
        
        self.point_number = -1
        
    def set_visited(self):
        
        self.visited = True
        
        [x, y] = self.point
        
        reserved_in_row[x][y] = ''
        reserved_in_column[y][x] = ''
        
        reserved_in_square[number_of_square([x, y])].remove(self.value)
        #self.square_number = -1
        
        all_visited = True
        
        for kid in self.parent.kids:
            all_visited *= kid.visited
            
        if all_visited:
            
            self.parent.set_visited()
            
        
        
    def get_posible_next_steps(self):
        
        #global reserved_in_row, reserved_in_column, Alpha
        
        Alpha_able = Alpha.copy() 
        
        [x, y] = self.point
        
        # находим какая ячейка следующая
        if self.parent != None:
            next_point_number = permitted_points.index(self.point) + 1
            if next_point_number >= len(permitted_points): 
                return []
            else:
                [x_next, y_next] = permitted_points[next_point_number]
            
        else:
            [x_next, y_next] = permitted_points[0]
        
        # фиктивно кладем
        if self.parent != None:
            reserved_in_row[x][y] = self.value
            reserved_in_column[y][x] = self.value

            reserved_in_square[number_of_square([x, y])].add(self.value)
            

        # составляем множество запрещенных значений
        disable_in_row = reserved_in_row[x_next].copy()
        disable_in_column = reserved_in_column[y_next].copy()
        
        disable_in_square = list(reserved_in_square[number_of_square([x_next, y_next])].copy())
    
        
        able = set(Alpha_able)
        
        disable = set(disable_in_row + disable_in_column + disable_in_square)
        
        # достаем то, что фиктивно положили
        if self.parent != None:
            reserved_in_row[x][y] = ''
            reserved_in_column[y][x] = ''

            reserved_in_square[number_of_square([x, y])].remove(self.value)

        
        return list(able - disable)   
    

def is_solution(cell):
    if v.point in permitted_points:
        if permitted_points.index(v.point) == len(permitted_points) - 1:
            return True
    return False

# init state ==============
[x_r, y_r] = permitted_points[0]

root = cell()


root.point = ( [x_r, y_r-1] if y_r != 0 or x_r == 0 else [x_r-1, n-1] )

root.value = ('' if y_r == 0 and x_r == 0 else reserved_in_row[root.point[0]][root.point[1]])

root.square_number = ( -1 if y_r == 0 and x_r == 0 else number_of_square([root.point[0], root.point[0]]) )
    
root.posible_steps = root.get_posible_next_steps()
        

tree_stack = [root]
# init state ==============

#point_number = 0
        
while len(tree_stack) != 0:
    
    v = tree_stack.pop()
    
#     print(v.value)
    
    [x, y] = v.point
    
    if v.parent != None: # если не корень
        
        reserved_in_row[x][y] = v.value
        reserved_in_column[y][x] = v.value
        
        reserved_in_square[number_of_square([x, y])].add(v.value)

    if is_solution(v):
        
        break 
    
    else:
        
        if (len(v.posible_steps)) == 0: # если у вершины нету возможных следующих шагов, то есть это лист,
            # то дотаем значение обратно и переходим к следующей итерации, на которой достаем из стека следующее
            # значение
            
            # если оказались в листе при том, что он не является решением, то нужно  вернуть сигнал к родителю
            v.set_visited()
        
        else: # если есть возможные следующие шаги, то инициализируем детей, 
            # перемешиваем и кладем в стек, чтобы потом обойти если понадобится

            # init kids for v
            for k in range(len(v.posible_steps)):

                kid = cell()
                kid.parent = v
                
                if v.parent != None:
                    next_point_number = permitted_points.index(v.point) + 1
                else:
                    next_point_number = 0 # kid.point = [0, 0]
                
                kid.point = permitted_points[next_point_number]
                    
                kid.square_number = number_of_square(kid.point)

                kid.value = v.posible_steps[k]

                kid.posible_steps = kid.get_posible_next_steps()

                kid.kids = []

                v.kids.append(kid)

            # lok for posible next steps
            posible_next_steps = v.kids.copy()
            rn.shuffle(posible_next_steps)

            # save posible next steps to check at next iteration
            tree_stack.extend(posible_next_steps)

            # если оказались в листе, то стек не расширится 
            # и на следующей итерации из него достанется следующее значение
            
#     print('==========')
#     values_in_stack = []
#     for node in tree_stack:
#         values_in_stack.append((node.value, node.posible_steps, node.point, node.square_number))
#     print(values_in_stack)
#     print(reserved_in_row)
#     print(reserved_in_column)
#     print()


for i in range(n):
    for j in range(n):
        print(reserved_in_row[i][j], end=' ')
    print()

    
```
