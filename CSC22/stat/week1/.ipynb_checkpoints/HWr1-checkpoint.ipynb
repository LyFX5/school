```python
#Задача 1. Напишите функцию, которая считает взвешенную корреляцию для выборки пар чисел.


def weighted_corr(sample: np.array, weights: np.array) -> float:
    # sample.shape = (n, 2)
    
    def weighted_mean(sample: np.array, weights: np.array):
        return np.dot(sample,weights)
    
    def weighted_std(sample: np.array, weights: np.array):
        M = weighted_mean(sample, weights)
        mean = M * np.ones_like(sample)
        difs = sample - mean
        W = np.diag(weights)
        s = np.dot(difs.T,np.dot(W,difs))
       
        return s**0.5
    
    X = sample[:,0]
    Y = sample[:,1]
    
    EX = weighted_mean(X, weights)
    EY = weighted_mean(Y, weights)
    
    DX = weighted_std(X, weights)
    DY = weighted_std(Y, weights)
    
    difsX = X - EX * np.ones_like(X)
    difsY = Y - EY * np.ones_like(Y)
    
    W = np.diag(weights)
    
    cor = np.dot(difsX.T,np.dot(W,difsY)) / (DX * DY)
    
    return cor
```
